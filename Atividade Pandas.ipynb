Atividade utilizando a biblioteca pandas, sobre a base de dados "Where in the world do people emit the most CO2?", que mostra a emissão per capita de CO2 em vários países.

## Importação
Importação da biblioteca pandas que será o foco dessa atividade

```python
import pandas as pd
```

## Leitura e Descrição dos dados
Criação do dataframe a partir do .csv.

```python
df = pd.read_csv("./datasets/co-emissions-per-capita.csv")
df.describe()
```

## Filtragem dos valores
A filtragem é feita utilizando dois países, Afeganistão e Brasil, além de filtrar todas as correspondências do ano de 2019.

```python
df_AFG = df[df["Entity"] == "Afghanistan"]
df_Brasil = df.loc[df['Entity'].str.contains("Brazil")]
df_2019 = df[df['Year'] == 2019]

df_AFG
```

```python
df_Brasil
```

```python
df_2019
```

## Agrupamento e Agregação
Calculando a média de emissão per capita de CO2 considerando o ano de 2019.

```python
total_emissao_2019 = df_2019.groupby("Year").agg({"Annual CO2 emissions (per capita)": "sum"})
qtd_emissao_2019 = df_2019.groupby("Year").agg({"Annual CO2 emissions (per capita)": "count"})
total_emissao_2019
```

```python
qtd_emissao_2019
```

```python
media_emissao_2019 = df_2019.groupby("Year").agg({"Annual CO2 emissions (per capita)": "mean"})
media_emissao_2019
```
